```python
# Read in data from both data sets:

df_billboard = pd.read_csv("billboard.csv")

df_audio = pd.read_csv("audio_features.csv")
```

```python
# we want to get rid of duplicates to avoid adding extra rows
df_audio = df_audio.drop_duplicates(subset=['song'])
                    
# defaults to left join
df_merge = df_billboard.set_index('song_id').join(df_audio.set_index('song_id'), rsuffix="_spotify")
```

```python
print("df_merge shape:", df_merge.shape)
df_merge.dtypes
```

```python
# drop features we wil not be using in our analyses

df_clean = df_merge.drop(columns = ['performer_spotify', # this is a duplicate
                                    'song_spotify', # this is a duplicate
                                    'spotify_track_preview_url', # not useful
                                    'url']) # not useful
```

```python
percent_missing = df_clean.isna().sum() * 100 / len(df_clean)

percent_missing_merge = pd.DataFrame(
    {'percent_missing': percent_missing})

print(percent_missing_merge)
```

```python
percent_missing_spotify = df_audio.isna().sum() * 100 / len(df_audio)

percent_missing_spotify = pd.DataFrame(
    {'percent_missing': percent_missing})

print(percent_missing_spotify)
```

```python
# filling NAs to be "debut", NAs represent the first time song has debuted on the Billboard charts

df_clean['previous_week_position'] = df_clean['previous_week_position'].fillna('debut')
```

```python
import ast
```

```python
df_clean['spotify_genre_clean'] = df_clean['spotify_genre'].fillna('[]')

df_clean['spotify_genre_clean'] = df_clean['spotify_genre_clean'].apply(lambda x : ast.literal_eval(x) if (x) else [])
```

```python
def mapping_genre(genres):
    if genres == []:
        return "NA"
    assigned_genre = "other"
    for genre in genres:
        if "country" in genre:
            return "country"
        elif "indie" in genre:
            return "indie"
        elif "edm" in genre:
            return "edm"
        elif "trap" in genre:
            return "trap"
        elif "rap" in genre:
            return "rap"
        elif "hip hop" in genre:
            return "hip hop"
        elif "r&b" in genre:
            return "r&b"
        elif "k-pop" == genre:
            return "k-pop"
        elif "jazz" in genre:
            return "jazz"
        elif "house" in genre:
            return "house"
        elif "folk" in genre:
            return "folk"
        elif "funk" in genre:
            return "funk"
        elif "soul" in genre:
            return "soul"
        elif "blues" in genre:
            return "blues"
        elif "punk" in genre:
            return "punk"
        # we want to see if we can find a more specific genre other than pop, but maybe we don't?
        elif "rock" in genre:
            assigned_genre = "rock"
        elif "pop" in genre:
            assigned_genre = "pop"
            
    return assigned_genre

df_clean['spotify_genre_clean'] = df_clean['spotify_genre_clean'].apply(mapping_genre)
```
